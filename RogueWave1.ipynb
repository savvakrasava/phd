```python
#packages

import pandas as pd
import os
import sys
import glob
import matplotlib.pyplot as plt
```

```python
#constants
    
path = 'data'
```

```python
#functions

def create_dataset(path):
    '''
    func for merge all data after simulations
    '''
    all_files = glob.glob(path+'/*.txt')
    temp = []
    
    for filename in all_files:
        df = pd.read_csv(filename, sep = '\t', header = None).assign(File = filename.split('.')[0])
        temp.append(df)
    
    return pd.concat(temp, axis = 0, ignore_index = True)
```

```python
#create dataframe from all files

df = create_dataset(path)
```

```python
#save data to .csv for lovely wife

df.to_csv('all_data.csv',index = False)
```
